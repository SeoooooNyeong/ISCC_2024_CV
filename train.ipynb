```python
from ultralytics import YOLO

model = YOLO("./runs/detect/train4/weights/best.pt")
```

```python
import cv2
%matplotlib inline
import matplotlib.pylab as plt
```

```python
test_image_path = "./images/s01340770.jpg"
# results = model(test_image_path, conf=0.01)
results = model.predict(test_image_path, conf=0.3)
image = cv2.imread(test_image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

results = results[0]
# 예측 결과에서 라벨 정보 추출 및 시각화
boxes = results.boxes  # 객체의 바운딩 박스
names = results.names  # 클래스 이름

# 바운딩 박스 그리기
for i in range(len(boxes)):
    class_index = int(boxes.cls[i])  # 클래스 인덱스
    class_name = names[class_index]  # 클래스 이름
    confidence = boxes.conf[i]  # 신뢰도 점수
    print("confidence", confidence)
    x1, y1, x2, y2 = map(int, boxes.xyxy[i])  # 바운딩 박스 좌표를 정수로 변환

    # 바운딩 박스와 라벨 텍스트 그리기
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
    label = f"{class_name}: {confidence:.2f}"
    fontScale = 1.0  # 글자 크기 조정
    thickness = 2  # 글자 두께 조정
    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, fontScale, (0, 255, 0), thickness)

# 이미지 출력
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

video_path = './test/KakaoTalk_20240627_174022564.mp4'
output_path = './test/medium_model/test6.mp4'

# 비디오 캡처 초기화
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

if not out.isOpened():
    print("Error: Could not open output video file.")
    cap.release()
    exit()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 프레임을 RGB로 변환
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # 객체 감지
    results = model.predict(image, conf=0.3)  
    results = results[0]
    boxes = results.boxes  # 객체의 바운딩 박스
    names = results.names  # 클래스 이름

    # 바운딩 박스 그리기
    for box in boxes:
        class_index = int(box.cls.item())  # 클래스 인덱스
        class_name = names[class_index]  # 클래스 이름
        confidence = box.conf.item()  # 신뢰도 점수
        print("confidence", confidence)
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # 바운딩 박스 좌표를 정수로 변환

        # 바운딩 박스와 라벨 텍스트 그리기
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        label = f"{class_name}: {confidence:.2f}"
        fontScale = 1.0  # 글자 크기 조정
        thickness = 2  # 글자 두께 조정
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, fontScale, (0, 255, 0), thickness)


    # 출력 비디오에 프레임 추가
    out.write(frame)

# 비디오 캡처 해제 및 출력 비디오 닫기
cap.release()
out.release()
cv2.destroyAllWindows()

print("Video processing complete. Saved as output.mp4")
```
