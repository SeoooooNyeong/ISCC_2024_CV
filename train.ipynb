```python
from ultralytics import YOLO

model = YOLO("./runs/detect/train4/weights/best.pt")
```

```python
import cv2
%matplotlib inline
import matplotlib.pylab as plt
```

```python
test_image_path = "./images/s01340770.jpg"
# results = model(test_image_path, conf=0.01)
results = model.predict(test_image_path, conf=0.3)
image = cv2.imread(test_image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

results = results[0]
# ì˜ˆì¸¡ ê²°ê³¼ì—ì„œ ë¼ë²¨ ì •ë³´ ì¶”ì¶œ ë° ì‹œê°í™”
boxes = results.boxes  # ê°ì²´ì˜ ë°”ìš´ë”© ë°•ìŠ¤
names = results.names  # í´ëž˜ìŠ¤ ì´ë¦„

# ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸°
for i in range(len(boxes)):
    class_index = int(boxes.cls[i])  # í´ëž˜ìŠ¤ ì¸ë±ìŠ¤
    class_name = names[class_index]  # í´ëž˜ìŠ¤ ì´ë¦„
    confidence = boxes.conf[i]  # ì‹ ë¢°ë„ ì ìˆ˜
    print("confidence", confidence)
    x1, y1, x2, y2 = map(int, boxes.xyxy[i])  # ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œë¥¼ ì •ìˆ˜ë¡œ ë³€í™˜

    # ë°”ìš´ë”© ë°•ìŠ¤ì™€ ë¼ë²¨ í…ìŠ¤íŠ¸ ê·¸ë¦¬ê¸°
    cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
    label = f"{class_name}: {confidence:.2f}"
    fontScale = 1.0  # ê¸€ìž í¬ê¸° ì¡°ì •
    thickness = 2  # ê¸€ìž ë‘ê»˜ ì¡°ì •
    cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, fontScale, (0, 255, 0), thickness)

# ì´ë¯¸ì§€ ì¶œë ¥
plt.figure(figsize=(10, 10))
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

video_path = './test/KakaoTalk_20240627_174022564.mp4'
output_path = './test/medium_model/test6.mp4'

# ë¹„ë””ì˜¤ ìº¡ì²˜ ì´ˆê¸°í™”
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

if not out.isOpened():
    print("Error: Could not open output video file.")
    cap.release()
    exit()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # í”„ë ˆìž„ì„ RGBë¡œ ë³€í™˜
    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # ê°ì²´ ê°ì§€
    results = model.predict(image, conf=0.3)  
    results = results[0]
    boxes = results.boxes  # ê°ì²´ì˜ ë°”ìš´ë”© ë°•ìŠ¤
    names = results.names  # í´ëž˜ìŠ¤ ì´ë¦„

    # ë°”ìš´ë”© ë°•ìŠ¤ ê·¸ë¦¬ê¸°
    for box in boxes:
        class_index = int(box.cls.item())  # í´ëž˜ìŠ¤ ì¸ë±ìŠ¤
        class_name = names[class_index]  # í´ëž˜ìŠ¤ ì´ë¦„
        confidence = box.conf.item()  # ì‹ ë¢°ë„ ì ìˆ˜
        print("confidence", confidence)
        x1, y1, x2, y2 = map(int, box.xyxy[0])  # ë°”ìš´ë”© ë°•ìŠ¤ ì¢Œí‘œë¥¼ ì •ìˆ˜ë¡œ ë³€í™˜

        # ë°”ìš´ë”© ë°•ìŠ¤ì™€ ë¼ë²¨ í…ìŠ¤íŠ¸ ê·¸ë¦¬ê¸°
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        label = f"{class_name}: {confidence:.2f}"
        fontScale = 1.0  # ê¸€ìž í¬ê¸° ì¡°ì •
        thickness = 2  # ê¸€ìž ë‘ê»˜ ì¡°ì •
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, fontScale, (0, 255, 0), thickness)


    # ì¶œë ¥ ë¹„ë””ì˜¤ì— í”„ë ˆìž„ ì¶”ê°€
    out.write(frame)

# ë¹„ë””ì˜¤ ìº¡ì²˜ í•´ì œ ë° ì¶œë ¥ ë¹„ë””ì˜¤ ë‹«ê¸°
cap.release()
out.release()
cv2.destroyAllWindows()

print("Video processing complete. Saved as output.mp4")
```
